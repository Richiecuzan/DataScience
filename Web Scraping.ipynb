```python
import requests as rq

url = "https://www.indeed.com/jobs?q=data%20science&l=Remote&vjk=b5e42183805170a5"
res = rq.get(url)

print(res)
```

```python
from bs4 import BeautifulSoup as soup  #import the libraries

jobs = soup(res.content) #get content from page
```

print(jobs)

```python
# get element by id

results = jobs.find(id="resultsCol")
results
```

```python
print(results)
```

```python
c_resuts1 = results.find_all("h2", class_="jobTitle")
job_titles = [t.text.strip() for t in c_resuts1] # get all job roles
```

```python
job_titles
```

```python
import pandas as pd

df = pd.DataFrame(job_titles, columns=["Job Title"])  # put the data in a dataframe

df.head()
```

```python
# coolect our data into a csv

df.to_csv("Indeed Remote Jobs First page.csv")
```
